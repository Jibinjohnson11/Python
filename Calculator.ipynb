```python
from tkinter import *

def click(num):
    a=txt.insert(END,num)
    
def clear():
    txt.delete(0,END)
    
def eva():
    equal()
    txt.insert(END,l)
    
def equal():
    while True:
        global l
        h= txt.get()
        b=str(h)
        d=eval(b)
        l=('=',d)
        return l
    
    
def repeat():
    try:
        eva()
    except:
        clear()
        txt.insert(END,'syntax error')
        
    
    
window = Tk()

window.title("calculator")
window.geometry('300x300')

txt = Entry(window,width=34)
txt.place(x=20,y=15)

btn1=Button(window,text=1,width=4,bg='gray',command=lambda:click(1))
btn1.place(x=20,y=60)
btn2=Button(window,text=2,width=4,bg='gray',command=lambda:click(2))
btn2.place(x=70,y=60)
btn3=Button(window,text=3,width=4,bg='gray',command=lambda:click(3))
btn3.place(x=120,y=60)
btn4=Button(window,text=4,width=4,bg='gray',command=lambda:click(4))
btn4.place(x=20,y=90)
btn5=Button(window,text=5,width=4,bg='gray',command=lambda:click(5))
btn5.place(x=70,y=90)
btn6=Button(window,text=6,width=4,bg='gray',command=lambda:click(6))
btn6.place(x=120,y=90)
btn7=Button(window,text=7,width=4,bg='gray',command=lambda:click(7))
btn7.place(x=20,y=120)
btn8=Button(window,text=8,width=4,bg='gray',command=lambda:click(8))
btn8.place(x=70,y=120)
btn9=Button(window,text=9,width=4,bg='gray',command=lambda:click(9))
btn9.place(x=120,y=120)
btn0=Button(window,text=0,width=4,bg='gray',command=lambda:click(0))
btn0.place(x=70,y=150)
btn_point=Button(window,text='x',width=4,bg='gray',command=lambda:txt.insert(END,'*'))
btn_point.place(x=120,y=150)
btn_clear=Button(window,text='C',width=4,bg='gray',command=lambda:clear())
btn_clear.place(x=20,y=150)
btn_eq=Button(window,text='=',width=4,bg='orange',command=repeat)
btn_eq.place(x=170,y=150)
btn_add=Button(window,text='+',width=4,bg='gray',command=lambda:txt.insert(END,'+'))
btn_add.place(x=170,y=120)
btn_mai=Button(window,text='-',width=4,bg='gray',command=lambda:txt.insert(END,'-'))
btn_mai.place(x=170,y=90)
btn_div=Button(window,text='/',width=4,bg='gray',command=lambda:txt.insert(END,'/'))
btn_div.place(x=170,y=60)


window.mainloop()
```
